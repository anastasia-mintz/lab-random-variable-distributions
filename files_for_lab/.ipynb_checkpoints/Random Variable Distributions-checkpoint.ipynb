```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
# Get the numerical variables from our dataset.

numericals = data.select_dtypes(np.number)
numericals.head()
```

```python
# Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

for each_column in numericals.columns:
 sns.displot(numericals[each_column])
```

```python
# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

def log_transform(x):
  x = np.log10(x)
  if np.isfinite(x):
    return x
  else:
    return 0

# Apply the function
numerical_log = numericals
for i in numerical_log.columns:
  numerical_log[i] = numerical_log[i].apply(log_transform)

for i in numerical_log.columns:
  sns.displot(data=numerical_log[i], bins = 200)
```
